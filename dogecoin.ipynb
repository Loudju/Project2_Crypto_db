```python
#Import dependancies
import pandas as pd
import datetime
```

```python
#Import CSV files      
dogecoin19= '../Project2_Crypto_db/dogecoin_full_data__4__2019.csv'
dogecoin20= '../Project2_Crypto_db/dogecoin_full_data__4__2020.csv'
dogecoin21= '../Project2_Crypto_db/dogecoin_full_data__4__2021.csv'
```

```python
#Read CSV files and move into dataframe
dogecoin19_df= pd.read_csv(dogecoin19)
dogecoin20_df= pd.read_csv(dogecoin20)
dogecoin21_df= pd.read_csv(dogecoin21)
```

```python
#Delete unnecessary columns for 2019 data
del dogecoin19_df['Unnamed: 0']
del dogecoin19_df['Asset_Name']
del dogecoin19_df['Weight']
del dogecoin19_df['Target']
dogecoin19_df.head()
```

```python
#Rename the columns in new data frame
dogecoin19_df=dogecoin19_df.rename(columns={'Asset_ID': "Crypto Currency Type", 'timestamp': 'Timestamp', 'Count': 'Number of Trades'})
dogecoin19_df
```

```python
#Format the columns accordingly
dogecoin19_df['Number of Trades'] = dogecoin19_df['Number of Trades'].map('{:,.0f}'.format)
dogecoin19_df['Open'] = dogecoin19_df['Open'].map('${:,.4f}'.format)
dogecoin19_df['High'] = dogecoin19_df['High'].map('${:,.4f}'.format)
dogecoin19_df['Low'] = dogecoin19_df['Low'].map('${:,.4f}'.format)
dogecoin19_df['Close'] = dogecoin19_df['Close'].map('${:,.4f}'.format)
dogecoin19_df['Volume'] = dogecoin19_df['Volume'].map('${:,.4f}'.format)
dogecoin19_df['VWAP'] = dogecoin19_df['VWAP'].map('${:,.4f}'.format)
dogecoin19_df.head()
```

```python
#Rewrite content in Crypto Currency Type
dogecoin_rename= {4:"Dogecoin"}

dogecoin19_df["Crypto Currency Type"]=dogecoin19_df["Crypto Currency Type"].map(dogecoin_rename)
dogecoin19_df
```

```python
#Insert input to change epoch time to readable time 
def epoch(inputdate):
    return datetime.datetime.fromtimestamp(inputdate).strftime('%Y-%m-%d %H:%M:%S')
```

```python
#Rename each column for ALL dataframes
timestamp= dogecoin19_df['Timestamp']
timestamp1= dogecoin20_df['timestamp']
timestamp2= dogecoin21_df['timestamp']
```

```python
#Input code to apply  new information to the designated column for all dataframes
dogecoin19_df['Date Time'] = dogecoin19_df['Timestamp'].apply(lambda x:epoch(x))
dogecoin20_df['Date Time'] = dogecoin20_df['timestamp'].apply(lambda x:epoch(x))
dogecoin21_df['Date Time'] = dogecoin21_df['timestamp'].apply(lambda x:epoch(x))
```

```python
#Read column
dogecoin19_df['Date Time']
```

```python
#Rename column for 2019 dataframe with new information
dogecoin19_df['Timestamp'] = dogecoin19_df['Date Time']
del dogecoin19_df['Timestamp']
dogecoin19_df.head()
```

```python
#Reorganize columns for 2019 dataframe
dogecoin19_df=dogecoin19_df[['Date Time', 'Crypto Currency Type', 'Number of Trades', 'Open', 'High', 'Low', 'Close', 'Volume', 'VWAP']]
dogecoin19_df.head()
```
