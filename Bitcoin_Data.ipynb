```python
import pandas as pd
```

```python
# Name of the CSV files we are working with
bitcoin19 = '../Project2_Crypto_db/bitcoin_full_data__1__2019.csv'
bitcoin20 = '../Project2_Crypto_db/bitcoin_full_data__1__2020.csv'
bitcoin21 = '../Project2_Crypto_db/bitcoin_full_data__1__2021.csv'
```

```python
# Read CSV files and move into DF
bitcoin19_df = pd.read_csv(bitcoin19)
bitcoin20_df = pd.read_csv(bitcoin20)
bitcoin21_df = pd.read_csv(bitcoin21)
```

```python
# Read Bitcoin 2019 Data
bitcoin19_df.head()
```

```python
# Delete columns Asset Name,Weight, and Unnamed:0
del bitcoin19_df['Unnamed: 0']
del bitcoin19_df['Asset_Name']
del bitcoin19_df['Weight']
del bitcoin19_df['Target']
bitcoin19_df.head()
```

```python
# Rename the columns in new data frame
bitcoin19_df = bitcoin19_df.rename(columns={'Asset_ID':'Crypto Currency Type','timestamp':'Timestamp','Count':'Number of Trades','Open':'Bitcoin Open','Close':'Bitcoin Close','High':'Bitcoin High','Low':'Bitcoin Low','Volume':'Bitcoin Volume','VWAP':'Bitcoin VWAP'})
bitcoin19_df

```

```python
# Format the columns accordingly
bitcoin19_df['Number of Trades'] = bitcoin19_df['Number of Trades'].map('{:,.0f}'.format)
bitcoin19_df['Bitcoin Open'] = bitcoin19_df['Bitcoin Open'].map('${:,.4f}'.format)
bitcoin19_df['Bitcoin High'] = bitcoin19_df['Bitcoin High'].map('${:,.4f}'.format)
bitcoin19_df['Bitcoin Low'] = bitcoin19_df['Bitcoin Low'].map('${:,.4f}'.format)
bitcoin19_df['Bitcoin Close'] = bitcoin19_df['Bitcoin Close'].map('${:,.4f}'.format)
bitcoin19_df['Bitcoin Volume'] = bitcoin19_df['Bitcoin Volume'].map('${:,.4f}'.format)
bitcoin19_df['Bitcoin VWAP'] = bitcoin19_df['Bitcoin VWAP'].map('${:,.4f}'.format)
bitcoin19_df.head()
```
