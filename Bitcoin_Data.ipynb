```python
import pandas as pd
```

```python
# Name of the CSV files we are working with
bitcoin19 = '../Project2_Crypto_db/bitcoin_full_data__1__2019.csv'
bitcoin20 = '../Project2_Crypto_db/bitcoin_full_data__1__2020.csv'
bitcoin21 = '../Project2_Crypto_db/bitcoin_full_data__1__2021.csv'
```

```python
# Read CSV files and move into DF
bitcoin19_df = pd.read_csv(bitcoin19)
bitcoin20_df = pd.read_csv(bitcoin20)
bitcoin21_df = pd.read_csv(bitcoin21)
```

Bitcoin 2019

```python
# Read Bitcoin 2019 Data
bitcoin19_df.head()
```

```python
# Delete columns Asset Name,Weight, and Unnamed:0
del bitcoin19_df['Unnamed: 0']
del bitcoin19_df['Asset_Name']
del bitcoin19_df['Weight']
del bitcoin19_df['Target']
bitcoin19_df.head()
```

```python
# Rename the columns in new data frame
bitcoin19_df = bitcoin19_df.rename(columns={'Asset_ID':'Crypto Currency Type','timestamp':'Timestamp','Count':'Number of Trades',})
bitcoin19_df

```

```python
# Format the columns accordingly
bitcoin19_df['Number of Trades'] = bitcoin19_df['Number of Trades'].map('{:,.0f}'.format)
bitcoin19_df['Open'] = bitcoin19_df['Open'].map('${:,.4f}'.format)
bitcoin19_df['High'] = bitcoin19_df['High'].map('${:,.4f}'.format)
bitcoin19_df['Low'] = bitcoin19_df['Low'].map('${:,.4f}'.format)
bitcoin19_df['Close'] = bitcoin19_df['Close'].map('${:,.4f}'.format)
bitcoin19_df['Volume'] = bitcoin19_df['Volume'].map('{:,.4f}'.format)
bitcoin19_df['VWAP'] = bitcoin19_df['VWAP'].map('${:,.4f}'.format)
bitcoin19_df.head()
```

```python
# Rewrite content in Crypto Currency Type
bitcoin_coin = {1:'Bitcoin'}
bitcoin19_df['Crypto Currency Type'] = bitcoin19_df['Crypto Currency Type'].map(bitcoin_coin)
bitcoin19_df.head()
```
