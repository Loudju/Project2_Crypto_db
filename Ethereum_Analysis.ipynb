```python
import pandas as pd
import datetime 
```

```python
# Ethereum CSV files
ethereum19 = '/Users/julian/Project2_Crypto_db/Project2_Crypto_db/Ethereum__full_data__2019.csv'
ethereum20 = '/Users/julian/Project2_Crypto_db/Project2_Crypto_db/Ethereum__full_data__2020.csv'
ethereum21 = '/Users/julian/Project2_Crypto_db/Project2_Crypto_db/Ethereum__full_data__2021.csv'
```

```python
# Read CSV files and move into DF 
ethereum19_df = pd.read_csv(ethereum19)
ethereum20_df = pd.read_csv(ethereum20)
ethereum21_df = pd.read_csv(ethereum21)
```

```python
# Read Ethereum 2019 Data
ethereum19_df.head()
```

```python
# Delete columns Asset Name, Weight, and Unnamed:0
del ethereum19_df['Unnamed: 0']
del ethereum19_df['Asset_Name']
del ethereum19_df['Weight']
del ethereum19_df['Target']
ethereum19_df.head()
```

```python
# Rename columns in new dataframe
ethereum19_df = ethereum19_df.rename(columns={'Asset_ID':'Crypto Currency Type','timestamp':'Timestamp','Count':'Number of Trades',})
ethereum19_df
```

```python
# Format the columns accordingly
ethereum19_df['Number of Trades'] = ethereum19_df['Number of Trades'].map('{:,.0f}'.format)
ethereum19_df['Open'] = ethereum19_df['Open'].map('${:,.4f}'.format)
ethereum19_df['High'] = ethereum19_df['High'].map('${:,.4f}'.format)
ethereum19_df['Low'] = ethereum19_df['Low'].map('${:,.4f}'.format)
ethereum19_df['Close'] = ethereum19_df['Close'].map('${:,.4f}'.format)
ethereum19_df['Volume'] = ethereum19_df['Volume'].map('${:,.4f}'.format)
ethereum19_df['VWAP'] = ethereum19_df['VWAP'].map('${:,.4f}'.format)
ethereum19_df.head()
```

```python
ethereum_coin = {6:'Ethereum'}
ethereum19_df['Crypto Currency Type'] = ethereum19_df['Crypto Currency Type'].map(ethereum_coin)
ethereum19_df.head()
```

```python
def epoch(inputdate):
    return datetime.datetime.fromtimestamp(inputdate).strftime('%Y-%m-%d %H:%M:%S')

timestamp = ethereum19_df['Timestamp']
ethereum19_df['Date Time'] = ethereum19_df['Timestamp'].apply(lambda x:epoch(x))
ethereum19_df['Date Time']
```

```python
ethereum19_df['Timestamp'] = ethereum19_df['Date Time']
del ethereum19_df['Timestamp']
ethereum19_df.head()
```

```python
ethereum19_df = ethereum19_df[['Date Time','Crypto Currency Type','Number of Trades','Open','High','Low','Close','Volume','VWAP']]
ethereum19_df.head()
```
